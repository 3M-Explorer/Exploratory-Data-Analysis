## From the histogram above, it was determined that the 'release_year' data shows a left-skewed distribution. Therefore, Tukey's method will be employed.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import MaxNLocator

# Tukey's Fences Method for Outlier Removal
def tukey_method(df, column):
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)

    # Compute the Interquartile Range (IQR)
    IQR = Q3 - Q1

    # Define the lower and upper bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Filter the DataFrame to remove outliers
    df_clean = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

    return df_clean, lower_bound, upper_bound

# Apply Tukey method to remove outliers
df_clean, lower_bound, upper_bound = tukey_method(df, 'release_year')

# Calculate the percentage of outliers removed
original_count = len(df)
cleaned_count = len(df_clean)
outliers_removed = original_count - cleaned_count
percentage_removed = (outliers_removed / original_count) * 100

# Display outlier bounds and percentage removed
print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
print(f"Outliers Removed: {outliers_removed} out of {original_count} rows")
print(f"Percentage of outliers removed: {percentage_removed:.2f}%")

# Create a comparison with box plots before and after outlier removal
plt.figure(figsize=(12, 6))

# Box plot before outlier removal
plt.subplot(1, 2, 1)
sns.boxplot(x=df['release_year'])
plt.title('Box Plot Before Outlier Removal')
plt.gca().xaxis.set_major_locator(MaxNLocator(integer=True))  # Ensure x-axis has integer ticks

# Box plot after outlier removal
plt.subplot(1, 2, 2)
sns.boxplot(x=df_clean['release_year'])
plt.title('Box Plot After Outlier Removal')
plt.gca().xaxis.set_major_locator(MaxNLocator(integer=True))  # Ensure x-axis has integer ticks

plt.tight_layout()
plt.show()
```
